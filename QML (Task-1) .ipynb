```python
import numpy as np
import matplotlib.pyplot as plt

print("\n" + "="*50)
print("TASK 1: BORN RULE - MEASUREMENT PROBABILITIES")
print("="*50)

def born_rule_probabilities(psi):
    """
    Calculate measurement probabilities using the Born rule:
    P(i) = |<i|ψ>|^2 = |ψ_i|^2
    """
    probabilities = np.abs(psi) ** 2
    return probabilities / np.sum(probabilities)  # Ensure normalization

# Define superposition states
psi_1 = np.array([1/np.sqrt(2), 1/np.sqrt(2)])  # |+⟩ state
psi_2 = np.array([1/np.sqrt(3), np.sqrt(2/3)])  # Custom normalized state

# Print states and probabilities
print("Superposition state 1: |ψ₁⟩ =", psi_1)
print("Measurement probabilities (|ψ₁⟩):", born_rule_probabilities(psi_1))
print("Superposition state 2: |ψ₂⟩ =", psi_2)
print("Measurement probabilities (|ψ₂⟩):", born_rule_probabilities(psi_2))

# Prepare labels and values for plotting
states = ['|0⟩', '|1⟩']
probs_1 = born_rule_probabilities(psi_1)
probs_2 = born_rule_probabilities(psi_2)

# Plotting
plt.figure(figsize=(10, 4))

# First subplot for ψ₁
plt.subplot(1, 5, 5)
plt.bar(states, probs_1, color='blue', alpha=0.7)
plt.title('State |ψ₁⟩ Measurement Probabilities')
plt.ylabel('Probability')
plt.ylim(0, 1)

# Second subplot for ψ₂
plt.subplot(1, 3, 1)
plt.bar(states, probs_2, color='yellow', alpha=0.7)
plt.title('State |ψ₂⟩ Measurement Probabilities')
plt.ylabel('Probability')
plt.ylim(0, 1)

plt.tight_layout()
plt.show()
```
